```python
import pandas as pd
import numpy as np

file_path = '/content/drive/My Drive/merged_2022_따릉이대여이력_수정.csv'
df = pd.read_csv(file_path, encoding='cp949')


df_filtered = df[((df['대여 자치구'] == '동작구') & (df['반납 자치구'] == '동작구')) |
                 ((df['대여 자치구'] == '관악구') & (df['반납 자치구'] == '관악구')) |
                 ((df['대여 자치구'] == '서초구') & (df['반납 자치구'] == '서초구')) |
                 ((df['대여 자치구'] == '강남구') & (df['반납 자치구'] == '강남구')) |
                 ((df['대여 자치구'] == '송파구') & (df['반납 자치구'] == '송파구'))]

df_filtered.to_csv('동작관악서초강남송파_2022_따릉이대여이력.csv', index=False, encoding='cp949')
```

```python
file_path2 = '/content/drive/My Drive/대여소정보.csv'
df_station = pd.read_csv(file_path2, encoding='utf-8')
df_station_name = df_station[['대여소번호', '보관소(대여소)명']].drop_duplicates()
```

```python
def return_max_combo(x):
    combo_counts = x.groupby(['대여 자치구', '대여 대여소번호', '반납대여소번호']).size().reset_index(name='빈도수')
    max_combo = combo_counts.sort_values(by='빈도수', ascending=False)
    return max_combo

def return_top_3(x, y):
    top_3 = pd.merge(x, df_station_name, left_on='대여 대여소번호', right_on='대여소번호', how='left')
    top_3 = pd.merge(top_3, df_station_name, left_on='반납대여소번호', right_on='대여소번호', how='left')
    top_3 = top_3.drop(columns=['대여소번호_x', '대여소번호_y'])
    top_3 = top_3.rename(columns={'보관소(대여소)명_x':'대여 대여소명', '보관소(대여소)명_y':'반납 대여소명', '위도_x':'대여소 위도', '경도_x':'대여소 경도', '위도_y':'반납소 위도', '경도_y':'반납소 경도'})
    top_3['백분율'] = (top_3['빈도수'] / len(y)) * 100
    return top_3
```

```python
#동작구
df_dongjak_diff = df_dongjak[(df_dongjak['대여 대여소번호'] != df_dongjak['반납대여소번호'])]
df_dongjak_top3 = return_top_3(return_max_combo(df_dongjak_diff), df_dongjak).head(3)
df_dongjak_top3
```

```python
#관악구
df_gwanak_diff = df_gwanak[(df_gwanak['대여 대여소번호'] != df_gwanak['반납대여소번호'])]
df_gwanak_top3 = return_top_3(return_max_combo(df_gwanak_diff), df_gwanak).head(3)
df_gwanak_top3
```
